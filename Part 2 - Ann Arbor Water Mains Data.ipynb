Here I will filter the data set in watermain_breaks_data to only consist of records from 1/1/2018 to 7/31/2023.

```python
# For this exercise, consider only data from 1/1/2018 to 7/31/2023. Assign these values.
start_date = pd.to_datetime('1/2018')
end_date = pd.to_datetime('7/2023')

# Convert Project Start Date to a datetime type
watermain_breaks_data['Project Start Date'] = pd.to_datetime(watermain_breaks_data['Project Start Date'])

# Create a month and year column
watermain_breaks_data['Project Start Month'] = watermain_breaks_data['Project Start Date'].dt.month
watermain_breaks_data['Project Start Year'] = watermain_breaks_data['Project Start Date'].dt.year

# Create function to format the date in a month/year format
def create_date_col(row):
    return f"{int(row['Project Start Month'])}/{int(row['Project Start Year'])}"

# Create a column with the month/year format by applying the function
watermain_breaks_data['Date'] = watermain_breaks_data.apply(create_date_col, axis = 1)

# Filter the data from 1/1/2018 to 7/31/2023
watermain_breaks_data = watermain_breaks_data[(pd.to_datetime(watermain_breaks_data['Date']) >= start_date) & (pd.to_datetime(watermain_breaks_data['Date']) <= end_date)]

# View the data
watermain_breaks_data.head()
```

## 1. Using “Watermain_Breaks_Reporting.csv”, compute the average number of watermain work orders for each month (based on project start start), aggregating across years (i.e., January, February, etc.). What month has the highest number of watermain breaks, on average?



```python
# Group the data to count the project start month and year
avg_monthly_work_orders = watermain_breaks_data.groupby(['Project Start Month', 'Project Start Year']).count()

# Group the data by month to find the average number of watermain work orders
avg_monthly_work_orders = avg_monthly_work_orders.groupby('Project Start Month').mean()

# View data
avg_monthly_work_orders['Work Order ID']
```

```python
# Find the month with the greatest average of watermain work orders
max_index = avg_monthly_work_orders['Work Order ID'].idxmax()
max_month = max_index

# Find the greatest average of watermain work orders
max_mean = avg_monthly_work_orders.loc[max_index]['Work Order ID']

print(f"On average, the month with the highest number of watermain work orders is the {max_month}th month, with an average of {max_mean}.")
```

## 1a. Plot the average number of watermain work orders by month in a line chart. Are there any identifiable trends? Is there any seasonality to the workorder requests?

```python
# Plot the average number of watermain work orders by month 

# Retrieve months
months = avg_monthly_work_orders.index

# Plot months as the x-axis, and the data line as the groupby product's averages
plt.plot(months, avg_monthly_work_orders['Work Order ID'], label = 'Average Number of Watermain Breaks Per Month', color = 'b', marker = 'o')

# Name the plot and add axes labels to describe the findings 
plt.title('Average Number of Watermain Work Orders by Month')
plt.xlabel('Months')
plt.ylabel('Number of Watermain Work Orders')

plt.show()
```

The most watermain work orders occur during the winter months of the year. This could be from the colder temperatures and frozen ground. As the water and ground freeze against the constricted ground, pipes can break. This can also occur in transition months such as October to November because the seasonal transition is a more sudden drop in temperature. This shifts and freezes the ground where the watermain's pipes are located. This explains the steep increase from October to December.

To calculate the average number of watermain work orders for each month, I had to group the data by the repective month and year in the data frame. I counted how many times there was watermain work order associated to the month and year. Then I grouped by the months and found those averages. To find the highest average number of of watermain breaks per month, I calculated the mean from the second groupby object. Then I used its maximum value to index the month. 

Overall, a lineplot was the best option to graph the data since we are measuring the changes of the variable over time. The x-axis in the line plot is equivalent to the months. The y-axis is the number of watermain work orders. I found the months from the latest groupby object's index. Its only column is the average number of work orders, so these are the points on the line plot. From there, I was able to find the trends and conclude that there is some seasonality to the number of work orders per month. 

# 2. Compute the cancel rate for watermain repairs (number of work orders cancelled divided by the total number of work orders) by month and year (i.e., 1/2018, 2/2018, etc.).


```python
# Replace the cancel rate codes from Y/N to 1/0 in order to calculate the mean
watermain_breaks_data['Cancel Rate'] = watermain_breaks_data['Canceled'].str.replace('N', '0').str.replace('Y', '1').astype(int)

# Calculate the average cancel rates per months and year
cancel_rates = pd.DataFrame(watermain_breaks_data.groupby(['Project Start Year', 'Project Start Month'])['Cancel Rate'].mean()).reset_index()
cancel_rates
```

## 2a. Plot the cancellation rate by month and year from 1/2018 through 7/2023. Are thereany identifiable trends? Is there any seasonality to the cancellation rates?


```python
# View the cancel rates data
cancel_rates
```

```python
# Create function to make dates in a month/year format
def create_date_col(row):
    return f"{int(row['Project Start Month'])}/{int(row['Project Start Year'])}"

# Create a column with the month/year format by applying the function
cancel_rates['Date'] = cancel_rates.apply(create_date_col, axis = 1)
```

```python
# Plot the cancel rate from 1/2018 to 7/2023
plt.figure(figsize=[20, 6])
plt.plot(cancel_rates['Date'], cancel_rates['Cancel Rate'], linestyle = '-')

# Name the plot and add axes labels to describe the findings 
plt.xlabel('Date')
plt.ylabel('Cancellation Rate (number cancelled/total)')
plt.title('Line Plot of Cancellation Rates by Month from January 2018 to July 2023')

# Adjust the x-axis values to prevent overlapping
plt.xticks(cancel_rates['Date'], rotation = 60, ha='right')

plt.legend()
plt.show()
```

To find the average nmber of watermains associated with a single work order, I merged the attributed and watermain breaks data on their work order id columns. I found that inner join worked best because it would merge the data without having lots of gaps or null values like the products of outer merges have. Then I wanted to see the total number of watermains for each wrok order. So, I grouped by the work order and counted the amount of facility IDs were related to each work order ID. The facility ID was the column that identified where the watermain was located, therefore distinguishing the unique watermain. After that, I calculated the averages for each work order and found the maximum number of watermains to around 1.2 for a singular work order. 

## 4. For each work order, compute the total length of watermain replaced as well as the time in days it took to complete the work order (the difference between “Actual Start” and “Actual Finish”). Create a scatterplot using these values. Is there any discernable relationship between these two measures?

Start by finding the days it took to complete the work orders.

```python
# Create a column for the number of days it took to complete a work order
merged_df['Days to Complete'] = (pd.to_datetime(merged_df['Actual Finish']) - pd.to_datetime(merged_df['Actual Start']))
```

```python
# drop all rows with neither 'Actual Start' and 'Actual Finish', drop all rows without finish dates
merged_df = merged_df.dropna(subset=['Actual Start', 'Actual Finish'], how = 'all')
merged_df = merged_df.dropna(subset=['Actual Finish'], how = 'all')

# Fill the null values in the new column to 1
merged_df['Days to Complete'] = merged_df['Days to Complete'].fillna(pd.Timedelta(days = 1))
```

```python
# Create a function to create fractional days 
def make_fractional_days(series):
    seconds = series.total_seconds()
    seconds_in_day = 24 * 60 * 60
    fractional = (seconds / seconds_in_day)
    return fractional 

# Create a new column for fractional days, use the function to do so
merged_df['Fractional Days to Complete'] = merged_df['Days to Complete'].apply(make_fractional_days)

# View data
merged_df['Fractional Days to Complete']
```

Now find the total length of the watermains were replaced.

```python
# Drop rows with null lengths
merged_df = merged_df.dropna(subset=['LENGTH'], how = 'all')
```

```python
# Group the data by each work order id, find the sum of fractional days and length
days_length_per_work_order = merged_df.groupby('WORKORDERI')[['Fractional Days to Complete', 'LENGTH']].sum()
days_length_per_work_order
```

```python
# Plot the number of days and length replaced for each work order
plt.figure(figsize=(8, 6))

# Define x and y values 
x = days_length_per_work_order['LENGTH']
y = days_length_per_work_order['Fractional Days to Complete']

# Plot the values
plt.scatter(x, y, alpha = 0.5)

# Name the plot and add axes labels to describe the findings 
plt.title('Scatter Plot of Days Took to Complete vs. Length Replaced of Each Work Order')
plt.xlabel('Length')
plt.ylabel('Days to Complete')

plt.grid(True)
plt.show()
```

The scatter plot shows the relationship betweem the days it took to complete a work order and the length that was replaced. In general, the amount of days it takes on average is around 1 day. I find that this is reasonable since many records in the merged data set did not have large differences between the dates the work order started and ended. There were a few outliers in both separate data sets, hence the outliers on the plot. There were a few occurences where the projects took many more days than the other, or a lot more of the pipe length was replaced. With this being said, it is fair to conclude that the work orders are completed with efficiency. Work orders usually take around a day to complete according to the data, even if the pipe length replaced varies. However, it is understanable that some points on the plot have more days if there is more length replaced.
